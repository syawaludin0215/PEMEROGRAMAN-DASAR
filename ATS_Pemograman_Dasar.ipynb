```python
suhu = float(input("Masukkan suhu air pendingin (°C): "))
if suhu <= 90:
    print("STATUS: Normal")
elif 90 < suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

Program ini meminta pengguna memasukkan nilai suhu air. Nilai tersebut kemudian dibandingkan menggunakan percabangan if, elif, dan else. Jika suhu berada pada atau di bawah 90°C, sistem menunjukkan kondisi normal. Jika suhu berada di atas 90°C tetapi kurang dari 100°C, sistem memberi peringatan bahwa suhu sedang meningkat dan perlu perhatian. Jika suhu mencapai atau melebihi 100°C, program menampilkan pesan “ALERT: SHUTDOWN” yang menunjukkan kondisi berbahaya sehingga sistem perlu dihentikan atau ditangani segera.

Masalah yang dapat diselesaikan program ini adalah pemantauan dan pengendalian suhu pada sistem pembangkit agar tetap berada dalam kondisi aman dan untuk mendeteksi kondisi darurat secara otomatis.

***suhu = float(input("Masukkan suhu air pendingin (°C): "))***

Fungsi baris ini:
Meminta pengguna memasukkan nilai suhu (dalam °C) melalui keyboard, lalu mengubahnya menjadi tipe data float agar bisa dibandingkan sebagai angka desimal.
Contoh: jika user mengetik 95.5, maka suhu = 95.5.

**if suhu <= 90:**
    print("STATUS: Normal")

Makna:
Jika suhu kurang dari atau sama dengan 90°C, sistem menampilkan pesan bahwa status pendingin normal.
Bagian ini adalah kondisi pertama (utama) yang dicek.

**elif 90 < suhu < 100:**
    print("STATUS: Warning")

Makna:
Jika suhu lebih dari 90°C tetapi masih di bawah 100°C, sistem menampilkan peringatan (Warning).
Kata elif berarti “else if”, yaitu kondisi alternatif jika syarat pertama tidak terpenuhi.

**else:**
    print("ALERT: SHUTDOWN")

Makna:
Jika suhu tidak memenuhi dua kondisi sebelumnya (berarti suhu ≥ 100°C), maka sistem menampilkan pesan “ALERT: SHUTDOWN” — artinya sistem harus dimatikan untuk keamanan.

```python
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

Program menyimpan data vibrasi dari 5 unit mesin dalam sebuah list. Loop for digunakan untuk memeriksa setiap nilai vibrasi berdasarkan indeksnya.
Jika nilai vibrasi lebih besar dari 2.0, program mencetak peringatan bahwa unit tersebut mengalami vibrasi tinggi.
Masalah yang diselesaikan:
Program ini membantu mendeteksi mesin yang berpotensi bermasalah karena vibrasi berlebih, sehingga tindakan perawatan dapat dilakukan lebih cepat untuk mencegah kerusakan.

```python
suhu_saat_ini = 60

while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini, "°C")
    suhu_saat_ini += 5


print("Suhu target tercapai:", suhu_saat_ini, "°C")
```

Penjelasan cara kerja program:

Program memulai suhu awal pada 60°C. Loop while akan terus berjalan selama kondisi suhu masih di bawah 85°C. Pada setiap iterasi, suhu dinaikkan sebesar 5°C dan nilai suhu tersebut dicetak.
Program ini menyelesaikan masalah pemantauan kenaikan suhu pada sistem pemanas sampai mencapai batas aman (85°C).

**suhu** _**saat**_ **ini** = 60

Artinya:
Program memulai proses dengan suhu awal sebesar 60°C.
Nilai ini disimpan dalam variabel suhu_saat_ini dan akan terus berubah selama proses berjalan.

**while suhu_saat_ini < 85:**

Artinya:
Kondisi perulangan (loop).
Selama suhu saat ini masih di bawah 85°C, perintah di dalam blok while akan terus dijalankan berulang-ulang.
Begitu suhu mencapai atau melebihi 85°C, perulangan berhenti.

**print**("Suhu saat ini:", suhu_saat_ini, "°C")

Artinya:
Menampilkan nilai suhu saat ini di setiap iterasi (setiap kali loop berjalan).

**suhu_saat_ini += 5**

Artinya:
Setiap kali loop berjalan, suhu akan naik 5°C.
Simbol += berarti “tambah dan simpan hasilnya ke variabel yang sama”.

```python
persentase = float(input("Masukkan persentase pemakaian beban (%): "))
if persentase <= 60:
    status = "Beban Ringan"
elif persentase <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"


print("Status beban listrik:", status)
```


Penjelasan cara kerja program:

Program meminta input kapasitas maksimum dan beban terpakai. Kemudian program menghitung persentase pemakaian dengan membagi beban terpakai terhadap kapasitas maksimum dan mengalikan hasilnya dengan 100. Nilai persentase ini digunakan dalam percabangan untuk menentukan kategori beban. Bila pemakaian ≤ 60%, statusnya “Beban Ringan”. Bila > 60% hingga ≤ 85%, statusnya “Beban Sedang”. Bila lebih dari 85%, program menampilkan “Beban Berat”.

Masalah yang dapat diselesaikan:

Program ini dapat digunakan untuk memantau tingkat pemakaian beban listrik sehingga operator dapat mengetahui apakah sistem berada pada kondisi aman, mendekati batas, atau berisiko kelebihan beban.

**persentase = float**(input("Masukkan persentase pemakaian beban (%): "))

Artinya:
Program meminta pengguna untuk memasukkan nilai persentase pemakaian beban listrik, misalnya 75.
Fungsi float() digunakan agar input dianggap sebagai bilangan desimal, bukan teks.

**if persentase** <= 60:
    status = "Beban Ringan"

Artinya:
Jika persentase beban kurang dari atau sama dengan 60%, maka statusnya adalah "Beban Ringan".
Ini menunjukkan sistem hanya menggunakan sedikit daya listrik.

**elif persentase **<= 85:
    status = "Beban Sedang"

Artinya:
Jika nilai lebih besar dari 60% tetapi masih ≤ 85%, maka statusnya "Beban Sedang".
elif berarti "else if", yaitu kondisi alternatif setelah kondisi pertama gagal.

**else**:
    status = "Beban Berat"

Artinya:
Jika tidak memenuhi dua kondisi sebelumnya (berarti lebih dari 85%), maka sistem memberi status "Beban Berat", menandakan konsumsi daya tinggi.

**print**("Status beban listrik:", status)

Artinya:
Menampilkan hasil akhir status sesuai nilai persentase yang dimasukkan.
